LINEAR REGRESSION

```python
# Hardcoded Linear Regression (One variable)
import numpy as np

# Training data
X = [1, 2, 3, 4, 5]
Y = [2, 4, 6, 8, 10]

# Step 1: Mean
x_mean = np.mean(X)
y_mean = np.mean(Y)

# Step 2: Calculate slope (m) and intercept (c)
numerator = sum((X[i] - x_mean) * (Y[i] - y_mean) for i in range(len(X)))
denominator = sum((X[i] - x_mean) ** 2 for i in range(len(X)))

m = numerator / denominator
c = y_mean - m * x_mean

print(f"Slope (m): {m}, Intercept (c): {c}")

# Prediction function
def predict(x):
    return m * x + c

# Predict for a new x
x_new = 6
y_pred = predict(x_new)
print(f"Prediction for x={x_new}: y={y_pred}")
```

```python
import numpy as np

# Data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([20, 40, 60, 80, 100])

# Parameters
m = 0
c = 0
alpha = 0.01  # learning rate
epochs = 1000
n = len(X)

# Gradient Descent Loop
for i in range(epochs):
    y_pred = m * X + c
    error = Y - y_pred

    dm = -(2/n) * sum(X * error)
    dc = -(2/n) * sum(error)

    m = m - alpha * dm
    c = c - alpha * dc

print(f"Slope: {m:.2f}, Intercept: {c:.2f}")
print(f"Predicted for x=6: {m*6 + c}")
```

EVALUATION METRIC

```python
import numpy as np

y = np.array([10, 20, 30])
y_pred = np.array([12, 18, 33])

# MAE
mae = np.mean(np.abs(y - y_pred))

# RMSE
rmse = np.sqrt(np.mean((y - y_pred) ** 2))

# R² Score
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r2 = 1 - (ss_res / ss_tot)

print("MAE:", mae)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python
import numpy as np

# Step 1: Data
X = np.array([
    [1200, 3, 10],
    [1500, 4, 5],
    [1000, 2, 15],
    [1800, 4, 8],
    [1700, 3, 6]
], dtype=float)

y = np.array([50, 65, 40, 70, 68], dtype=float)

# Step 2: Normalize features (optional, helps with learning speed)
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Step 3: Initialize weights and bias
w = np.zeros(X.shape[1])  # [w1, w2, w3]
c = 0

# Step 4: Gradient Descent Parameters
lr = 0.01
epochs = 1000
n = len(y)

# Step 5: Training Loop
for i in range(epochs):
    y_pred = np.dot(X, w) + c
    error = y - y_pred
    
    # Gradients
    dw = -2/n * np.dot(X.T, error)
    dc = -2/n * np.sum(error)

    # Update weights
    w = w - lr * dw
    c = c - lr * dc

    if i % 100 == 0:
        mse = np.mean(error ** 2)
        print(f"Epoch {i}: MSE = {mse:.4f}")

# Step 6: Final Weights
print("\nLearned weights (w1, w2, w3):", w)
print("Learned bias (c):", c)

# Step 7: Prediction
# Predict price for new house: [1600, 3, 7]
x_new = np.array([1600, 3, 7])
x_new = (x_new - X.mean(axis=0)) / X.std(axis=0)  # normalize like training data
y_new_pred = np.dot(x_new, w) + c
print("Predicted price:", y_new_pred)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data
data = {
    'Area': [1200, 1500, 1000, 1800, 1700],
    'Bedrooms': [3, 4, 2, 4, 3],
    'Age': [10, 5, 15, 8, 6],
    'Price': [50, 65, 40, 70, 68]
}
df = pd.DataFrame(data)

# Features and Target
X = df[['Area', 'Bedrooms', 'Age']]  # 3 features
y = df['Price']                      # Target

# Train Model
model = LinearRegression()
model.fit(X, y)

# Coefficients
print("Weights (w1, w2, w3):", model.coef_)
print("Intercept (b):", model.intercept_)

# Predict for a new house
new_house = [[1600, 3, 7]]  # Area=1600, 3 bedrooms, 7 years old
price = model.predict(new_house)
print("Predicted Price (₹ Lakhs):", price[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Step 1: Sample data (clearly non-linear)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 8, 18, 32, 50])  # Not a straight line!

# Step 2: Create Polynomial Regression model (degree 2 or 3)
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X, y)

# Step 3: Predict new values
x_test = np.linspace(1, 5, 100).reshape(-1, 1)
y_pred = model.predict(x_test)

# Step 4: Plot
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(x_test, y_pred, color='red', label='Polynomial Fit')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 5, 10, 17, 26])

# Step 2: Try polynomial degrees 1 to 4
degrees = [1, 2, 3, 4]
colors = ['red', 'green', 'orange', 'purple']

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Original Data')

# Step 3: Loop through each degree
for i, degree in enumerate(degrees):
    # Step 3.1: Create polynomial features
    poly = PolynomialFeatures(degree=degree)
    x_poly = poly.fit_transform(x)

    # Step 3.2: Fit model
    model = LinearRegression()
    model.fit(x_poly, y)

    # Step 3.3: Predict y
    y_pred = model.predict(x_poly)

    # Step 3.4: Calculate cross-validation RMSE
    cv_scores = cross_val_score(model, x_poly, y, scoring='neg_mean_squared_error', cv=3)
    rmse = np.sqrt(-cv_scores.mean())
    r2 = r2_score(y, y_pred)

    print(f"Degree {degree}:")
    print(f"  RMSE = {rmse:.3f}")
    print(f"  R² Score = {r2:.3f}")

    # Step 3.5: Plot curve
    x_line = np.linspace(1, 5, 100).reshape(-1, 1)
    x_line_poly = poly.transform(x_line)
    y_line_pred = model.predict(x_line_poly)
    plt.plot(x_line, y_line_pred, color=colors[i], label=f'Degree {degree}')

# Final plot
plt.title("Polynomial Regression Curve Comparison")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression


X,y = make_regression(n_samples = 100,n_features = 1,noise = 15,random_state = 42)

y = y.reshape(-1,1)
m = X.shape[0]

X_b = np.c_[np.ones((m,1)),X]

theta = np.array([[2.0],[3.0]])

learning_rate = 0.1
n_iter = 100

for _ in range(n_iter):
    y_pred = X_b.dot(theta)
    gradient = (2/m)*X_b.T.dot(y_pred-y)
    theta -= learning_rate*gradient
    





plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue")
plt.plot(X, X_b.dot(theta), color="green")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()


```
